```python
d='Nature is the creation of natural material and physical world on earth independent of the human activities. The beauty and richness of nature is diverse and massive. Nature provides us with several delicacies to enjoy and is our life support system. Our basic needs like food, shelter, water and air are all the gifts from nature. Humans have a strong connection and interrelation with nature. Escaping away in the nature heals the mind and body and soothes our soul. Irrespective of what nature provides us it is also our moral and social responsibility to nurture nature.'
```

```python
d
```

```python
import string
a="".join([i for i in d if i not in [string.punctuation]])
a
```

```python
b = a.split()
b
```

```python
keys=[]         
# creating keys for dictionary
for k in b:
    if k not in keys:
        keys.append(k)
keys
```

```python
values=[]       
# creating values for dictionary
for x in keys:
    values.append(b.count(x))
    
values
```

```python
d1 = { k:v for (k,v) in zip(keys, values)}    # creating the dictionary using zip
d1
```

```python
import pandas as pd
df=pd.DataFrame({'word':list(d1.keys()),'no of occurences':d1.values()})  # creating the data frame
df
```
