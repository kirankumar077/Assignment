```python
import string
def remove_punc(str1):
    a = str.maketrans('', '', string.punctuation)
    str2 = str1.translate(a)
    return str2
str1 = "Nature is the ultimate source of our living. Both living and non-living things include nature, and everyone is interdependent, which helps maintain the ecosystem"
str2 = remove_punc(str1)
print(str2)
```

```python
def word_split(str2):
    str2 = str2.strip()
    L1 = str2.split()   
    return L1
str2 = word_split(str1)
str2
```

```python
import string
def word_dict(L1):
    d1 = {}
    for word in L1:
        # If the word is already in the dictionary, increment its count by 1
        if word in d1:
            d1[word] += 1
        # If the word is not in the dictionary, add it with a count of 1
        else:
            d1[word] = 1
    
    return d1
result = word_dict(str2)
print(result)
```

```python
import pandas as pd
def word_freq_table(d1):
    word_freq_list = []
    for word, count in d1.items():
        word_freq_list.append((word, count))
    df = pd.DataFrame(word_freq_list, columns=["word", "No of occurrence"])
    return df
result1= word_freq_table(result)
result1


```
